```python
#类和对象
#类：抽象的，唯一的，总体的
#对象：基于类创建的，基于一个类，可以创建很多对象，实在的，

#类和对象：属性，方法
#属性：变量
#方法：函数，代码块

```

```python
#  计算机内存  ##Mankind属性number、方法##   ##zhangSan属性、方法##   ##liSi属性、方法##
class Mankind:
    
    #在这里创建的属性，只属于类Mankind，不属于任何对象
    number = 0
    
    #定义属性
    def __init__(self,name,height,weight):
        Mankind.number = Mankind.number + 1
        self.name = name
        self.height = height
        self.weight = weight
        
    #定义方法,可以传参数，也能调用属性
    def eat(self, time):
        print("Time is ", time,".", self.name, "can eat!")
        
print(Mankind.number)
        
zhangSan = Mankind("zhangSan", 175, 65)

print(zhangSan.name)
print(zhangSan.height)
print(zhangSan.weight)

zhangSan.eat("12:00")

print("==================")
# 新的对象
liSi = Mankind("liSi", 185, 75)

print(liSi.name)
print(liSi.height)
print(liSi.weight)

liSi.eat("18:00")

#不存在覆盖，因为实例化类之后，对象就是新的存储位置。
print(zhangSan.name)

print(Mankind.number)
```

```python
#!/usr/bin/python3

class Fruit:
    number = 0
	def __init__(self,ID, name, place):
		Fruit.number += 1
		self.ID = ID
		self.name = name
		self.place = place

	def show(self):
		print(self.ID, self.name, self.place)

class Supermaket:
	def __init__(self,ID,name,fruits = []):
		self.ID = ID
		self.name = name
		self.fruits = fruits

	def addFruit(self, fruit):
		self.fruits.append(fruit)
        
    #把所有的水果都展示出来
    def showFruits(self):
        #遍历self.fruits，并调用Fruit的showFruit方法
        for i in self.fruits:
	        i.show()



a = Fruit(1,'apple', 'shangdong')
b = Fruit(2, 'mango', 'hainan')
c = Fruit(3, 'watermelon', 'hainan')

supermaket = Supermaket(1,'s')
supermaket.addFruit(a)
supermaket.addFruit(b)
supermaket.addFruit(c)

supermaket.showFruits()
```

```python
#!/usr/bin/python3
# 水仙花数
for i in range(100, 1000):
    sum = 0
    for j in list(str(i)):
        sum += pow(int(j), 3)
    if sum == i:
        print(i)


# 教师学生
class Teacher:
    def __init__(self, id):
        self.id = id
        self.course = []

    def choose(self, cou):
        self.course.append(cou)
        cou.addtea(self.id)
        # print('老师选课成功')

    def show(self):
        print('老师工号:', self.id, '您教的课有：', [i.name for i in self.course])

class Student:
    def __init__(self, id, name):
        self.id = id
        self.name = name
        self.course = []

    def choose(self, cou):
        self.course.append(cou)
        cou.addstu(self.id)
        # print('学生选课成功')

    def show(self):
        print('学生学号:', self.id, '你选的课有：', [i.name for i in self.course])

class Course:
    def __init__(self, id, name):
        self.id = id
        self.name = name
        self.tea = []
        self.stu = []

    def addstu(self, s):
        self.stu.append(s)

    def addtea(self, t):
        self.tea.append(t)
    def show(self):
        print('教', self.name, '课的老师有：', self.tea)
        print('选', self.name, '课的学生有：', self.stu)


if __name__ == '__main__':
    c = Course(1, '数学')
    c1 = Course(2, '语文')
    c2 = Course(3, '英语')
    c.show()

    t = Teacher(11)
    t.choose(c)
    t.choose(c1)
    t.show()

    s = Student(111, 'zsw')
    s.choose(c)
    s.choose(c2)
    s.show()

    s1 = Student(222, 'wqy')
    s1.choose(c)
    s1.choose(c1)
    s1.show()

    c.show()
```
