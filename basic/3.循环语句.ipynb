```python
#循环语句

for i in range(0,9):  # [0,9) == 0,1,2,...,8 默认step=1
    print(i)
```

```python
for i in range(0,9,2):  # [0,9) step=2
    print(i)
```

```python
for i in range(9):  # [默认0,9) step=1
    print(i)
```

```python
# 对容器中的元素进行遍历
list1 = [1,2,3,'a','b','c']

for i in list1: #set,tuple,dictionary
    print(i)


#for i in range(0,len(list1),2):
#    print(list1[i])
```

```python
# 判断+循环语句
a = 10
while(a < 20): #相当于if做了个判断
    a = a + 2
    print(a)
```

```python
#练习
#100以内的质数打印出来
for i in range(2,101):
    flag = True
    for j in range(2,i):
        if i % j == 0:
            flag = False
    if flag:
        print(i)
```

```python
for i in range(2,101):
    j=2
    for j in range(2,i):
        if(i%j==0):
            break
        else:
            print(i)
```
