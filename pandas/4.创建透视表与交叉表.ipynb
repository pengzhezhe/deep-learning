```python
import numpy as np
import pandas as pd

detail = pd.read_csv("./data/meal_order_detail1.csv",encoding="utf8")
```

```python
# pivot_table 函数
# 使用order_id作为透视表索引制作透视表 默认求平均值
detailPivot=pd.pivot_table(detail[['order_id','counts','amounts']],index='order_id')
print(detailPivot.head())
```

```python
detailPivot=pd.pivot_table(detail[['order_id','counts','amounts']],index='order_id',aggfunc=np.sum)
print(detailPivot.head())
```

```python
detailPivot=pd.pivot_table(detail[['order_id','dishes_name','counts','amounts']],index=['order_id','dishes_name'],aggfunc=np.sum)
print(detailPivot)
```

```python
# 指定菜品名称为列分组键的透视表 index索引本质就是行标签
detailPivot=pd.pivot_table(detail[['order_id','dishes_name','counts','amounts']],index='order_id',columns='dishes_name',aggfunc=np.sum)
print(detailPivot)
```

```python
# 指定某些列制作透视表
detailPivot=pd.pivot_table(detail[['order_id','dishes_name','counts','amounts']],index='order_id',values='counts',aggfunc=np.sum)
print(detailPivot)
```

```python
#对透视表中的缺失值进行填充
detailPivot=pd.pivot_table(detail[['order_id','dishes_name','counts','amounts']],index='order_id',columns='dishes_name',aggfunc=np.sum,fill_value=0)
print(detailPivot)
```

```python
#在透视表中添加汇总数据，表示总数
detailPivot=pd.pivot_table(detail[['order_id','dishes_name','counts','amounts']],index='order_id',columns='dishes_name',aggfunc=np.sum,fill_value=0,margins=True)
print(detailPivot)
```

```python
# 使用crosstab函数创建交叉表
detailCross=pd.crosstab(index=detail['order'],columns=detail['dishes_name'],values=detail['counts'],aggfunc=np.sum)
print(detailCross)
```
